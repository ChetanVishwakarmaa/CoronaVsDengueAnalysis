```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
corona_df = pd.read_csv('Latest Covid-19 India Status.csv')
dengue_df = pd.read_csv('RS_Session_255_AU_242.2-1.csv')
```

```python
corona_df.sort_values(by='Total Cases',ascending=False)
```

```python
dengue_df = dengue_df.iloc[:,[0,1,8,9] ]
```

```python
dengue_df
```

# Least affect states from Corona and Dengue

```python
w = 0.6
corona_df = corona_df.sort_values(by='Total Cases')
y=corona_df.iloc[0:11,1]
bar1=np.arange(len(y))
plt.bar(bar1,y,w)
x_labels = corona_df.iloc[0:11, 0]
plt.xticks(bar1, x_labels, rotation=45, ha='right')
plt.xlabel('States')
plt.ylabel('Total Cases')
plt.title('y Total Cases of corona')
plt.tight_layout()
plt.show()
```

```python
w = 0.6
dengue_df = dengue_df.sort_values(by='2021 (Prov. till 21st Nov) - Cases')
y= dengue_df.iloc[0:11,2]
bar1=np.arange(len(y))
plt.bar(bar1,y,w,color='orange')
x_labels = dengue_df.iloc[0:11, 1]
plt.xticks(bar1, x_labels, rotation=45, ha='right')
plt.xlabel('States')
plt.ylabel('Total Cases')
plt.title('Total Cases of dengue')
plt.tight_layout()
plt.show()
```

#Mostnumber of recovery in States

```python
corona_df = corona_df.sort_values(by='Discharged', ascending=False)
y=corona_df.iloc[0:11,3]
bar1=np.arange(len(y))
plt.bar(bar1,y,w)
x_labels = corona_df.iloc[0:11, 0]
plt.xticks(bar1, x_labels, rotation=45, ha='right')
plt.xlabel('States')
plt.ylabel('Total Discharged in Millions ')
plt.title(' Total Discharged by Corona virus')
plt.tight_layout()
plt.show()
```

```python
dengue_df['discharged'] = dengue_df['2021 (Prov. till 21st Nov) - Cases'] - dengue_df['2021 (Prov. till 21st Nov) - Deaths']
dengue_df = dengue_df.sort_values(by='discharged',ascending= False)
y= dengue_df.iloc[0:11,4]
bar1=np.arange(len(y))
plt.bar(bar1,y,w,color='orange')
x_labels = dengue_df.iloc[1:12, 1]
plt.xticks(bar1, x_labels, rotation=45, ha='right')
plt.xlabel('States')
plt.ylabel('Total Discharged')
plt.title('Total Discharged of dengue')
plt.tight_layout()
plt.show()
```

#States with highest Death Ratio

```python
corona_df = corona_df.sort_values(by='Death Ratio', ascending=False)
y=corona_df.iloc[0:11,7]
bar1=np.arange(len(y))
plt.bar(bar1,y,w)
x_labels = corona_df.iloc[0:11, 0]
plt.xticks(bar1, x_labels, rotation=45, ha='right')
plt.xlabel('States')
plt.ylabel('Death Ratio ')
plt.title(' Death Ratio of Corona virus')
plt.tight_layout()
plt.show()
```

```python
dengue_df['Death Ratio'] = (dengue_df['2021 (Prov. till 21st Nov) - Deaths']/dengue_df['2021 (Prov. till 21st Nov) - Cases'])*100
dengue_df = dengue_df.sort_values(by='Death Ratio',ascending= False)
y= dengue_df.iloc[0:11,5]
bar1=np.arange(len(y))
plt.bar(bar1,y,w,color='orange')
x_labels = dengue_df.iloc[1:12, 1]
plt.xticks(bar1, x_labels, rotation=45, ha='right')
plt.xlabel('States')
plt.ylabel('Death Ratio')
plt.title('Death Ratio of dengue')
plt.tight_layout()
plt.show()
```
